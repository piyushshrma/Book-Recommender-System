## Popularity Based Recommender System

```python
ratings_w_name = ratings.merge(books,on='ISBN')
```

```python
num_rating_df = ratings_w_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_w_name.groupby('Book-Title').agg({'Book-Rating': 'mean'}).reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df)
```

```python
popularity_df
```

```python
# Filter DataFrame
popularity_df = popularity_df[popularity_df['num_ratings'] >= 250]
popularity_df
```

```python
popularity_df = popularity_df.sort_values('avg_rating',ascending=False).head(50)
```

```python
popularity_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')
```

```python
popularity_df=popularity_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

```python
popularity_df
```

## Colaborative Filtering Based Recommender System

```python
ratings_w_name
```

```python
x=ratings_w_name.groupby('User-ID').count()['Book-Rating']>200
padhe_likhe_users = x[x].index
```

```python
filtered_ratings = ratings_w_name[ratings_w_name['User-ID'].isin(padhe_likhe_users)]
```

```python
filtered_ratings
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)].shape
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt  = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.shape
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

## Vector ko implement karenge

```python
from sklearn.metrics.pairwise import cosine_similarity as cs
```

```python
similarity_scores = cs(pt)
```

```python
similarity_scores.shape
```

```python

def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('Message in a Bottle')
```

```python
recommend('The Notebook')
```
